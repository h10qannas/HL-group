```python
import pandas as pd

# Load your dataset
df = pd.read_csv('/content/drive/MyDrive/Shabbek/shabbek_full_dataset.csv')

def get_smart_suggestions(user_id, num_suggestions=5):
    # 1. Identify the current user's role
    user_info = df[df['User_ID'] == user_id]
    if user_info.empty:
        return "User not found."

    user_role = user_info.iloc[0]['Role']

    # 2. Define the Business Logic Map
    # Who does this specific user need to meet?
    role_map = {
        'Investor': ['Entrepreneur', 'Startup'],
        'Entrepreneur': ['Investor', 'Service Provider'],
        'Service Provider': ['Startup', 'Entrepreneur'],
        'Startup': ['Service Provider', 'Investor']
    }

    # Get the list of target roles for the current user
    target_roles = role_map.get(user_role, [])

    # 3. Filter the dataset for those roles
    # We find people who are in the target roles AND are not the current user
    potential_matches = df[df['Role'].isin(target_roles) & (df['User_ID'] != user_id)]

    # 4. Sort by Points (Highest reputation first)
    suggestions = potential_matches.sort_values(by='Points', ascending=False)

    # 5. Return the results
    return suggestions[['User_ID', 'Full_Name', 'Role', 'Points']].head(num_suggestions)

```

```python
# Test it out
# If User 1 is an Entrepreneur, they should see Investors and Service Providers
print(get_smart_suggestions(user_id=1))
```
